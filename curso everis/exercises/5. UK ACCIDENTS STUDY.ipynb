#### **3.2. Identify NaN**

* We need the following columns to have always information: 'Location_Easting_OSGR', 'Location_Northing_OSGR', 'Longitude' and 'Latitude'

Create a new column with value 1 if all they have value and 0 if not.

```python
file[['Location_Easting_OSGR', 'Location_Northing_OSGR', 'Longitude' ]].isnull().any()
```

```python
def f(row):
    if np.isnan(row['Location_Easting_OSGR']) & np.isnan(row['Location_Northing_OSGR']) & np.isnan(row['Longitude']):
        return 1
    else:
        return 0
```

```python
file['dumi']= file[['Location_Easting_OSGR', 'Location_Northing_OSGR', 'Longitude' ]].apply(f, axis=1)
```

* How many rows are 0 for the new column?

```python
len(file[file['dumi']==0])
```

* Store in a new Dataframe rows with value 1 in the new column

```python
df = file[file['dumi']==1]
df.head()
```
